```python
import math

def compute_shares(datasets, tau, beta):
    # datasets: dict[name] = {"N":..., "L":...}
    W = {}
    for k, v in datasets.items():
        N, L = v["N"], v["L"]
        W[k] = (N ** tau) * ((math.log(L + 1)) ** beta if beta != 0 else 1.0)
    Z = sum(W.values())
    P = {k: W[k] / Z for k in W}

    # expected token share under token-budget stopping
    denom = sum(P[k] * datasets[k]["L"] for k in datasets)
    token_share = {k: (P[k] * datasets[k]["L"]) / denom for k in datasets}
    return token_share

# ==== 你在这里填统计量（示例数值请替换成你的真实统计） ====
datasets = {
    "GAME": {"N": 1e8, "L": 600},      # L 用你定义的均值或P50/P90
    "SWE-PRO": {"N": 732, "L": 6000},
    # 建议把其它子集也补上，否则分母不完整：
    "LGC": {"N": 1e6, "L": 800},
    "LGC-v2": {"N": 5e7, "L": 900},
    "CDE": {"N": 1e4, "L": 1200},
    "PRINT": {"N": 1e4, "L": 400},
    "Affine:abd": {"N": 1e4, "L": 900},
}

strategies = {
    "Proportional": (1.0, 0.0),
    "Temperature (tau=0.5)": (0.5, 0.0),
    "LADS (tau=0.7,beta=0.5)": (0.7, 0.5),
}

for name, (tau, beta) in strategies.items():
    ts = compute_shares(datasets, tau, beta)
    print(name, "GAME:", ts.get("GAME"), "SWE-PRO:", ts.get("SWE-PRO"))
```
